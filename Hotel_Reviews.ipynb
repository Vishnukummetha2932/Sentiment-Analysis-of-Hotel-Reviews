```python
import pandas as pd
from textblob import TextBlob

df = pd.read_csv('Hotel_Reviews.csv')

def get_sentiment(text):
    # Converting the text into string format in case of nulls values
    text = str(text)
    # Getting polarity score between(-1 to 1)
    polarity = TextBlob(text).sentiment.polarity
    # Classify the text based on polarity value
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['reviews.text'].apply(get_sentiment)

sentiment_counts = df['Sentiment'].value_counts()

print(sentiment_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    colors=['green', 'red', 'gray'],
    startangle=180
)

plt.title("Sentiment Analysis of Hotel Reviews")
plt.axis('equal')
plt.tight_layout()

legend_labels = [f"{label}: {count}" for label, count in zip(sentiment_counts.index, sentiment_counts)]
plt.legend(legend_labels, title="Sentiment Counts", loc="best")

plt.show()
```
